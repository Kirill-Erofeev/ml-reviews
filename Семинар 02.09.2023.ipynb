# Функции для парсинга HTML-файлов

```python
def improvement(html: bs4.element.Tag, cls: str):
    return html.find(class_=cls).get_text().strip()

   
def get_soup_dict(**kwargs) -> dict[str, bs4.BeautifulSoup]:
    soup_dict = {}
    for key, value in kwargs.items():
        with open(fr'.\{value}', 'r', encoding='UTF-8') as f:
            soup_dict[key] = BeautifulSoup(f.read(), 'html.parser')
    return soup_dict


def get_data_frame_dict(soup_dict: dict[str, bs4.BeautifulSoup]) -> dict[str, list]:
    scores, reviews, dates, authors = [], [], [], []
    for key, value in soup_dict.items():
        blocks = value.find_all(class_='lenta-card')
        for html in blocks:
            if html.find(class_='lenta-card__mymark'):
                dates.append(improvement(html, 'lenta-card__date').replace('\xa0', ' '))
                scores.append(improvement(html, 'lenta-card__mymark'))
                reviews.append(improvement(html, 'lenta-card__text without-readmore'))
                authors.append(key)
    return {
        'datetime': dates,
        'review': reviews,
        'score': scores,
        'author': authors,
    }
```

# Класс, унаследованный от pd.DataFrame для обработки и визуализации датафрейма

```python
# class SubclassedSeries(pd.Series):
#     @property
#     def _constructor(self):
#         return SubclassedSeries
    
#     @property
#     def _constructor_expanddim(self):
#         return SubclassedDataFrame



# class SubclassedDataFrame(pd.DataFrame):        
#     @property
#     def _constructor(self):
#         return SubclassedDataFrame
    
#     @property
#     def _constructor_sliced(self):
#         return SubclassedSeries

class SubclassedDataFrame(pd.DataFrame):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        
    def lowering_classes_by_ratios(
        self,
        class_ratios={
            '0': 0.045,
            '1': 0.01,
            '2': 0.045,
            '3': 0.1,
            '4': 0.25,
            '5': 0.55,
        }
    ) -> 'SubclassedDataFrame':
        balanced_data_frame = pd.DataFrame()
        for class_label, class_ratio in class_ratios.items():
            class_data = self[self['score'] == int(class_label)]
            if class_data.shape[0] > 0:
                sample_count = int(self.shape[0] * class_ratio)
                if class_data.shape[0] < sample_count:
                    sampled_data = class_data
                else:
                    sampled_data = class_data.sample(n=sample_count, replace=True)
                balanced_data_frame = pd.concat([balanced_data_frame, sampled_data])
        balanced_data_frame = balanced_data_frame.reset_index(drop=True)
        balanced_data_frame.__class__ = SubclassedDataFrame
        return balanced_data_frame
        
    def raising_the_mironitarian_classes(self) -> 'SubclassedDataFrame':
        balanced_data_frame = pd.DataFrame()
        class_ratios = self['score'].value_counts().to_dict()
        max_class_count = max(class_ratios.values())
        for class_label, class_count in class_ratios.items():
            class_data_frame = self[self['score'] == int(class_label)]
            if (max_class_count % class_count) / class_count > 0.5:
                class_data_frame = pd.concat(
                    [class_data_frame] * (max_class_count // class_count + 1),
                    axis=0
                )
            else:
                class_data_frame = pd.concat(
                    [class_data_frame] * (max_class_count // class_count),
                    axis=0
                )
            balanced_data_frame = pd.concat([balanced_data_frame, class_data_frame])
        balanced_data_frame = balanced_data_frame.reset_index(drop=True)
        balanced_data_frame.__class__ = SubclassedDataFrame
        return balanced_data_frame

    def lowering_majority_classes(self) -> 'SubclassedDataFrame':
        min_class_count = self['score'].value_counts().min()
        grouped = self.groupby('score')
        balanced_data_frame = grouped.apply(lambda x: x.sample(min_class_count))
        balanced_data_frame = balanced_data_frame.reset_index(drop=True)
        balanced_data_frame.__class__ = SubclassedDataFrame
        return balanced_data_frame
    
    def processing(self) -> 'SubclassedDataFrame':
        warnings.filterwarnings('ignore')
        self = self[self['score'].str.len() < 3]
        self = self.reset_index(drop=True)
        self['score'] = self['score'].astype('int64')
        self['datetime'] = self['datetime'] \
            .apply(lambda x: re.search(r'^[0-9]{1,2}', x)
            .group()) \
            .replace(
                {
                     '1': '01',
                     '2': '02',
                     '3': '03',
                     '4': '04',
                     '5': '05',
                     '6': '06',
                     '7': '07',
                     '8': '08',
                     '9': '09',
                }
            ) + \
            self['datetime'].apply(lambda x: re.search(r'[а-я]+', x)
                .group()) \
                .replace(
                    {
                        'января': '.01.',
                        'февраля': '.02.',
                        'марта': '.03.',
                        'апреля': '.04.',
                        'мая': '.05.',
                        'июня': '.06.',
                        'июля': '.07.',
                        'августа': '.08.',
                        'сентября': '.09.',
                        'октября': '.10.',
                        'ноября': '.11.',
                        'декабря': '.12.',
                    }
                ) + \
            self['datetime'].apply(lambda x: re.search(r'[2][0][0-9]{2}', x)
                .group())
        self = self.rename(columns={'datetime': 'date'})
        self.__class__ = SubclassedDataFrame
        return self
    
    def visualization_by_years(self) -> None:
        plt.figure(figsize=(15, 7))
        plt.title('Распределение отзывов по годам')
        plt.grid(color='black')
        plt.bar(
            self['date'].apply(lambda x: re.search(r'[0-9]{4}', x).group())
                .value_counts()
                .index,
            self['date'].apply(lambda x: re.search(r'[0-9]{4}', x).group())
                .value_counts(),
            color='lime'
        );
           
    def visualization_by_author(self) -> None:
        plt.figure(figsize=(15, 7))
        plt.title('Средние оценки по авторам')
        plt.grid(color='black')
        plt.bar(
            self.groupby('author').mean()
                .index,
            self.groupby('author').mean()
                .values
                .ravel(),
            color='aqua',
            width=0.1
        );

    def visualization_by_reviews(self) -> None:
        plt.figure(figsize=(15, 7))
        plt.title('Распределение отзывов по оценкам')
        plt.pie(
            self['score'].value_counts().values,
            autopct='%1.1f%%',
            colors=
            [
                'aqua',
                'violet',
                'yellow',
                'lime',
                'orange',
                'red',
            ]
        )
        plt.legend(self['score'].value_counts().index);
```

# Класс для машинного обучения

```python
class MachineLearning:
    def __init__(self, data_frame):
        self.data_frame = data_frame
    
    def preprocessing(self, list_of_reviews: list[str]) -> list[str] | None:
        morph_vocab = MorphVocab()
        list_of_processed_reviews = []
        for review in list_of_reviews:
            temporary_list = []
            processed_review = re.sub('[^а-яА-Я]', ' ', review)
            list_of_words = nltk.word_tokenize(processed_review)
            for word in list_of_words:
                temporary_list.append([_.normal_form for _ in morph_vocab(word)][0])
#             list_of_words = [_ for _ in list_of_words if _ not in stopwords.words('russian')]
            processed_review = ' '.join(temporary_list)
            list_of_processed_reviews.append(processed_review)
        if self is None:
            self.data_frame['review'] = list_of_processed_reviews
        else:
            return list_of_processed_reviews
    
    def learning_by_vectorizer(self, list_of_processed_reviews: list[str]) -> (
        list[list[int]],
        list[list[int]],
        list[list[int]],
        list[list[int]],
    ):
        count_vectorizer = CountVectorizer()
        matrix = count_vectorizer.fit_transform(
            self.data_frame['review'].values
        ).toarray()
        X = matrix
        y = self.data_frame['score'].values
        X_train, X_test, y_train, y_test = train_test_split(
            X,
            y,
            test_size=0.33,
            random_state=1
        )
        return (X_train, X_test, y_train, y_test)
    
    def learning_by_tfidf(self) -> (
        list[list[int]],
        list[list[int]],
        list[list[int]],
        list[list[int]],
    ):
        tfidf_vectorizer = TfidfVectorizer()
        matrix = tfidf_vectorizer.fit_transform(
            self.data_frame['review'].values
        )
        X = matrix
        y = self.data_frame['score'].values
        X_train, X_test, y_train, y_test = train_test_split(
            X,
            y,
            test_size=0.33,
            random_state=1
        )
        return (X_train, X_test, y_train, y_test)
    
    def logistic_regression(
        self,
        X_train: list[list[int]],
        X_test: list[list[int]],
        y_train: list[list[int]],
        y_test: list[list[int]],
    ) -> list[float]:
        lr = LogisticRegression()
        lr.fit(X_train, y_train)
        y_pred = lr.predict(X_test)
#         cv_results = cross_validate(lr,
#                                     X_test,
#                                     y_test,
#                                     cv=3,
#                                     scoring=('accuracy', 'f1', 'recall', 'precision'),
#                                     return_train_score=True,)
#         for key in cv_results:
#             print(f'key = {key}, m = {cv_results[key].mean()}')
#         print(cv_results.keys())

#         print(f'accuracy_score = {accuracy_score(y_test, y_pred):.3f}')
#         print(f'''f1_score = {f1_score(y_test, y_pred, average='weighted'):.3f}''')
#         print(f'''recall_score = {recall_score(y_test, y_pred, average='weighted'):.3f}''')
#         print(f'''precision_score = {precision_score(y_test, y_pred, average='weighted'):.3f}''')
        plt.title('Матрица ошибок LogisticRegression')
        sns.heatmap(confusion_matrix(y_test, y_pred), annot=True);
        return [
            accuracy_score(y_test, y_pred),
            f1_score(y_test, y_pred, average='weighted'),
            recall_score(y_test, y_pred, average='weighted'),
            precision_score(y_test, y_pred, average='weighted')
        ]

    def random_forest_classifier(
        self,
        X_train: list[list[int]],
        X_test: list[list[int]],
        y_train: list[list[int]],
        y_test: list[list[int]],
    ) -> list[float]: 
        rfc = RandomForestClassifier()
        rfc.fit(X_train, y_train)
        y_pred = rfc.predict(X_test)
        plt.title('Матрица ошибок RandomForestClassifier')
        sns.heatmap(confusion_matrix(y_test, y_pred), annot=True);
        return [
            accuracy_score(y_test, y_pred),
            f1_score(y_test, y_pred, average='weighted'),
            recall_score(y_test, y_pred, average='weighted'),
            precision_score(y_test, y_pred, average='weighted')
        ]
        
    def k_neighbors_classifier(
        self,
        X_train: list[list[int]],
        X_test: list[list[int]],
        y_train: list[list[int]],
        y_test: list[list[int]],
    ) -> list[float]:  
        knc = KNeighborsClassifier()
        knc.fit(X_train, y_train)
        y_pred = knc.predict(X_test)
        plt.title('Матрица ошибок KNeighborsClassifier')
        sns.heatmap(confusion_matrix(y_test, y_pred), annot=True);
        return [
            accuracy_score(y_test, y_pred),
            f1_score(y_test, y_pred, average='weighted'),
            recall_score(y_test, y_pred, average='weighted'),
            precision_score(y_test, y_pred, average='weighted')
        ]
        
    def ridge_classifier(
        self,
        X_train: list[list[int]],
        X_test: list[list[int]],
        y_train: list[list[int]],
        y_test: list[list[int]],
    ) -> list[float]:     
        rc = RidgeClassifier()
        rc.fit(X_train, y_train)
        y_pred = rc.predict(X_test)
        plt.title('Матрица ошибок RidgeClassifier')
        sns.heatmap(confusion_matrix(y_test, y_pred), annot=True);
        return [
            accuracy_score(y_test, y_pred),
            f1_score(y_test, y_pred, average='weighted'),
            recall_score(y_test, y_pred, average='weighted'),
            precision_score(y_test, y_pred, average='weighted'),
        ]
        
    def decision_tree_classifier(
        self,
        X_train: list[list[int]],
        X_test: list[list[int]],
        y_train: list[list[int]],
        y_test: list[list[int]],
    ) -> list[float]: 
        dtc = DecisionTreeClassifier()
        dtc.fit(X_train, y_train)
        y_pred = dtc.predict(X_test)
        plt.title('Матрица ошибок DecisionTreeClassifier')
        sns.heatmap(confusion_matrix(y_test, y_pred), annot=True);
        return [
            accuracy_score(y_test, y_pred),
            f1_score(y_test, y_pred, average='weighted'),
            recall_score(y_test, y_pred, average='weighted'),
            precision_score(y_test, y_pred, average='weighted')
        ]
    
    def predict(self, list_of_reviews) -> int:
        list_of_processed_reviews = self.preprocessing(list_of_reviews)
        tfidf_vectorizer = TfidfVectorizer()
        matrix = tfidf_vectorizer.fit_transform(
            self.data_frame['review'].values
        )
        review = tfidf_vectorizer.transform(
            list_of_processed_reviews
        )
        X = matrix
        y = self.data_frame['score'].values
        X_train, X_test, y_train, y_test = train_test_split(
            X,
            y,
            test_size=0.33,
            random_state=1
        )
        lr = LogisticRegression()
        lr.fit(X_train, y_train)
        ty = lr.predict(X_test)
        y_pred = lr.predict(review)
        print(y_pred)
```

```python
import re
import time
import requests
import warnings
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup



class SubclassedSeries(pd.Series):
    @property
    def _constructor(self):
        return SubclassedSeries

    @property
    def _constructor_expanddim(self):
        return SubclassedDataFrame

    
    
class SubclassedDataFrame(pd.DataFrame):        
    @property
    def _constructor(self):
        return SubclassedDataFrame

    @property
    def _constructor_sliced(self):
        return SubclassedSeries

    def processing(self, data_frame):
#         warnings.filterwarnings('ignore')
        data_frame = data_frame[data_frame['score'].str.len() < 3]
        data_frame['score'] = data_frame['score'].astype('int64')

        data_frame['date'] = data_frame['datetime'] \
            .apply(lambda x: re.search(r'^[0-9]{1,2}', x)
            .group()) \
            .replace(
                {
                     '1': '01',
                     '2': '02',
                     '3': '03',
                     '4': '04',
                     '5': '05',
                     '6': '06',
                     '7': '07',
                     '8': '08',
                     '9': '09',
                }
            ) + \
            data_frame['datetime'].apply(lambda x: re.search(r'[а-я]+', x)
            .group()) \
            .replace(
                {
                    'января': '.01.',
                    'февраля': '.02.',
                    'марта': '.03.',
                    'апреля': '.04.',
                    'мая': '.05.',
                    'июня': '.06.',
                    'июля': '.07.',
                    'августа': '.08.',
                    'сентября': '.09.',
                    'октября': '.10.',
                    'ноября': '.11.',
                    'декабря': '.12.'
                }
            ) + \
            data_frame['datetime'].apply(lambda x: re.search(r'[2][0][0-9]{2}', x)
            .group())
        
        return data_frame
        
        
    def visualization_by_author(self, data_frame):
        plt.figure(figsize=(15, 7))
        plt.title('Средние оценки по авторам')
        plt.grid(color='black')
        plt.bar(
            data_frame.groupby('author').mean()
                .index,
            data_frame.groupby('author').mean()
                .values
                .ravel(),
            color='darkgreen',
            width=0.1
        );
    
    def visualization_by_reviews(self, data_frame):
        plt.figure(figsize=(15, 7))
        plt.title('Распределение отзывов по оценкам')
        plt.pie(
            data_frame['score'].value_counts(),
            autopct='%1.1f%%',
            colors=[
                'aqua',
                'violet',
                'yellow',
                'lime',
                'orange',
                'red',
            ]
        )
        plt.legend(data_frame['score'].unique())
    

df = SubclassedDataFrame(dic)
df = df.processing(df)
df
# df.visualization_by_years(df)
# df.visualization_by_author(df)
# df.visualization_by_reviews(df)
```
